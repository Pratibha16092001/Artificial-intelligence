# Matplotlib Library

plotting library for pyhton pl and its numerical extension numpy.

```python
import matplotlib.pyplot as plt

%matplotlib inline   
#No need to write plt.show()
```

```python
import numpy as np
```

```python
x=np.arange(1,10)
y=np.arange(11,20)
```

# Scatter plot

```python
plt.scatter(x,y,c='red')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
```

# plt Plot

```python
y=x*x
```

```python
plt.plot(x,y,c='red',linestyle='--',markersize=12)
```

# Creating subplots

```python
# plt.subplot(no. of rows, no of column,position)
plt.subplot(2,2,1)
plt.plot(x,y,'blue')
plt.subplot(2,2,3)
plt.plot(x,y,'red')
plt.subplot(2,2,2)
plt.plot(x,y,'green')
plt.subplot(2,2,4)
plt.plot(x,y,'black')
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
```

```python
y=np.sin(x)
plt.subplot(3,2,1)
plt.plot(x,y,'blue')
plt.title('sine wave')

y=np.cos(x)
plt.subplot(3,2,2)
plt.plot(x,y,'red')
plt.title('cosine wave')

y=np.tan(x)
plt.subplot(3,2,3)
plt.plot(x,y,'green')
plt.title('tangent wave')

y=1/np.sin(x)
plt.subplot(3,2,4)
plt.plot(x,y,'black')
plt.title('cosec wave')

y=1/np.cos(x)
plt.subplot(3,2,5)
plt.plot(x,y,'yellow')
plt.title('sec wave')

y=1/np.tan(x)
plt.subplot(3,2,6)
plt.plot(x,y,'brown')
plt.title('cot wave')
```

# Bar plot

```python
x=[1,2,3,4]
y=[11,9,3,6]

x1=[19,10,4,7]
y1=[2,5,7,4]

plt.bar(x,y)
plt.bar(x1,y1,color='green')
plt.title('Bar Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

# Histogram

```python
a=np.array([11,34,65,23,56,78,65,89,45])
plt.hist(a)
plt.title('Histogram')
```

# Box plot using matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
# vert- vertical look, patch_artist-patch onit
```

# Pie chart

```python
# data to plot
labels=['python','ruby','cpp','java']
explode=[0.2,0,0,0]   #explode 1st slice
sizes=[14,35,25,65]
colors=['green','yellow','red','blue']

plt.pie(sizes,labels=labels,explode=explode,colors=colors , autopct='%1.1f%%',shadow=False)
plt.axis('equal')
```
