# How to create series in pandas

```python
#create series from list

l1=[1,2,3,4,5]

type(l1)

s1=pd.Series(l1)

s1
```

```python
s2=pd.Series(['hi','hello','how','are','you'])

s2
```

```python
age=[20,21,22,23,24,25]
s3=pd.Series(age,index=['a','b','c','d','e','f'])
print(s3)
```

```python
# How to create series using numpy array

arr=np.array(['e','d','u','n','e','t'])
s4=pd.Series(arr)
s4
```

```python
#how to create series from dictionary
```

```python
week_dict={
    'day1':'mon',
    'day2':'tues',
    'day3':'wed',
    'day4':'thurs'
}
week_dict
s5=pd.Series(week_dict)
s5
```

```python
#how to create series from tuple
t1=[1,2,3,4,5]
s6=pd.Series(t1)
s6
```

# How to create dataframe in pandas

```python
#How to create data frame from dictionary

weather_data={
    'day':['1-jan','2-jan','3-jan','4-jan','5-jan'],
    'temp':[45,67,46,38,56],
    'windspeed':[12,13,34,24,45],
    'event':['sunny','rain','cloudy','snow','sunny']
}

df1=pd.DataFrame(weather_data)
df1
```
