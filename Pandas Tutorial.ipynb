# Pandas Tutorial

### Pandas is an open source.Easy to use data structure and data analysis tools for the python programming lang.

```python
# Import pandas library
import pandas as pd
import numpy as np
```

```python
# To create dataframe by arange() 
# index- Rows, columns-columns
df=pd.DataFrame(np.arange(0,20).reshape(5,4),index=['row1','row2','row3','row4','row5'],columns=['column1','column2','column3','column4'])
```

```python
df
```

```python
# Converts dataframe to CSV file
df.to_csv('Test1.csv')
```

```python
# Accessing the elements
# 1- .loc  2.- .iloc
df.loc['row1']
```

```python
type(df.loc['row1'])
```

```python
# Access elements upto row3 and column2 
# 1st arg-row, 2nd arg-column
df.iloc[0:3,0:2]
```

```python
#Elemnts from row4 and column3
df.iloc[:4,:3]
```

```python
# Convert dataframes into array
df.iloc[:4,:3].values
```

```python
df.iloc[:4,:3].value_counts()
```

```python
df.iloc[:4,:3].values.shape
```

```python
df['column1'].unique()
```

```python
# Access multiple columns
df[['column1','column2']]
```

# CSV

```python
from io import StringIO, BytesIO
```

```python
data=('col1,col2,col3\n'
     'x,y,1\n'
     'a,b,2\n'
     'c,d,3')
```

```python
type(data)
```

```python
# If you want specific columns
df=pd.read_csv(StringIO(data),usecols=['col1','col3'])
```

```python
df
```

```python
# Convert dataframes into csv file
df.to_csv('test.csv')
```

```python
data1=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11,12')
```

```python
data1
```

```python
# Whenn you give datatype by your choice
df=pd.read_csv(StringIO(data1),dtype={'b':float,'c':object,'d':int})
```

```python
df
```

```python
# Check the datatype
df.dtypes
```

```python
# Make any column as index
data1=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11,12')
```

```python
pd.read_csv(StringIO(data1),index_col=1)
```
