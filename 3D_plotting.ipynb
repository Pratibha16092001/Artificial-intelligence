# pie_chart

```python
import matplotlib.pyplot as plt
```

```python
product=['mobile','laptops','desktops','tablets','smartphones']
product_sales_in_million=[35,100,40,60,80]
plt.pie(product_sales_in_million,labels=product,shadow=True,startangle=90,autopct='%1.2f%%')
```

# increase size of circle

```python
product=['mobile','laptops','desktops','tablets','smartphones']
product_sales_in_million=[35,100,40,60,80]
plt.pie(product_sales_in_million,labels=product,shadow=True,startangle=90,autopct='%1.2f%%',radius=2)
```

# visualization with numpy

```python
#plot sine wave using numpy and matpltlib
#linspace=define numbers
# arrange= define steps size

import numpy as np
x=np.arange(0,10,0.1)
y=np.sin(x)

print(x)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('sin(x)')
plt.title('sin graph')
```

```python
import numpy as np
x=np.arange(0,10,0.5)
y=np.sin(x)

plt.bar(x,y)
plt.xlabel('x values')
plt.ylabel('sin(x)')
plt.title('sin graph')
plt.show()
```

# 3D plotting

```python
from mpl_toolkits import mplot3d
import numpy as np
from matplotlib import pyplot as plt
```

```python
# how to plot empty 3d plot

%matplotlib notebook
fig= plt.figure()
ax=plt.axes(projection='3d')
```

```python
x=np.linspace(0,15,1000)
y= np.sin(x)
z=np.cos(x)
print(x)
print(y)
print(z)
```

```python
#HOW TO GENERATE 3D LINE PLOT
#%matplotlib notebook
fig= plt.figure()
ax=plt.axes(projection='3d')
ax.plot3D(x,y,z)
```

# HOW TO GENERATE 3D SCATTER PLOT

```python
#HOW TO GENERATE 3D LINE PLOT
x=np.linspace(0,15,100)
y= np.sin(x)
z=np.cos(x)
print(x)
print()
print(y)
print()
print(z)

%matplotlib notebook
fig= plt.figure()
ax=plt.axes(projection='3d')
ax.scatter3D(x,y,z)
```
