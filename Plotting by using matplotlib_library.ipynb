# matplotlib

```python
from matplotlib import pyplot as plt
```

# Simple line plot

```python
x=[1,4,6]
y=[3,2,5]

plt.plot(x,y)

plt.title("Simple line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Scatter line plot

# plot sine wave using numpy and matplotlib

```python
import numpy as np
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
print(x)
print(y)
plt.plot(x,y)
plt.title("sine wave")
plt.xlabel("x values")
plt.ylabel("sin(x) values")
```

# cos wave

```python
y=np.cos(x)
print(x)
print(y)
plt.plot(x,y)
plt.grid()
plt.title("cosine wave")
plt.xlabel("x values")
plt.ylabel("cos(x) values")
```

# tan wave

```python
y=np.tan(x)
print(x)
print(y)
plt.plot(x,y)
plt.title("tan wave")
plt.xlabel("x values")
plt.ylabel("tan(x) values")
```

# cosec wave

```python
y=1/np.sin(x)
print(x)
print(y)
plt.plot(x,y)
plt.title("cosec wave")
plt.xlabel("x values")
plt.ylabel("cosec(x) values")
```

# sec wave

```python
y=1/np.cos(x)
print(x)
print(y)
plt.plot(x,y)
plt.title("sec wave")
plt.xlabel("x values")
plt.ylabel("sec(x) values")
```

# cot wave

```python
y=1/np.tan(x)
print(x)
print(y)
plt.plot(x,y)
plt.grid()
plt.title("cot wave")
plt.xlabel("x values")
plt.ylabel("cot(x) values")
```

```python
a=np.arange(0,10,0.1)
b=np.sin(a)
c=np.cos(a)

plt.subplot(1,2,1)
plt.plot(a,b)

plt.subplot(1,2,2)
plt.plot(a,c)
```

# scatter plot

```python
days=[1,2,3,4,5,6]
max_temp=[45,46,47,49,50,54]

plt.xlabel=('days')
plt.ylabel('max_temp')
plt.scatter(days,max_temp)
plt.show()
```

```python
days=[1,2,3,4,5,6]
max_temp=[45,46,47,49,50,54]
min_temp=[23,24,27,29,31,36]

plt.xlabel=('days')
plt.ylabel('temperature')
plt.scatter(days,max_temp,label='max_temp')
plt.scatter(days,min_temp,label='min_temp')
plt.grid()
plt.title('days vs temp')
plt.legend(loc='upper right')
plt.show()
```

```python
##scatter plot sine wave using numpy & matplotlib
fig,ax=plt.subplots()
x=np.arange(0,10,0.1)
y=np.sin(x)

ax.plot(x,y)
ax.annotate('local maxima',xy=(1.7,1),xytext=(4,1.00),arrowprops=dict(facecolor='green'))
ax.annotate('local maxima',xy=(4.7,-1),xytext=(3.6,0.00),arrowprops=dict(facecolor='black'))
```
