# Pandas plotting

```python
import pandas as pd
```

```python
dict={
    'unemployment rate':[6.1,5.8,5.6,5.3,5.0,4.9,4.6,4.5],
    'stock_price_index':[1500,1525,1534,1560,1570,1585,1595,1600]
}

df=pd.DataFrame(dict)
df
```

```python
df.plot(x="unemployment rate",y="stock_price_index",kind="line",grid='true')
```

```python
df.plot(x="unemployment rate",y="stock_price_index",kind="scatter",grid='true')
```

```python
df.plot(x="unemployment rate",y="stock_price_index",kind="bar",grid='true')
```

# Bar graph

```python
data={
    'task':[300,500,700]
}
data
```

```python
df=pd.DataFrame(data,columns=['task'],index=['pending_tasks','progress_tasks','completed_tasks'])
df
```

```python
df.plot.pie(y='task')
```

```python
df.plot.pie(y='task',figsize=(15,5),autopct='%1.2f%%',startangle=0)
```
